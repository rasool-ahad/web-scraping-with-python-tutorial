# beatiful soup

### load libraries

```python
import requests 
from bs4 import BeautifulSoup as bs
```

## load our first page

```python
# load the website
r = requests.get("https://keithgalli.github.io/web-scraping/example.html")
# convert to beatiful object
soup = bs(r.content)
# print out our html
print(soup.prettify()) # or just print(soup)
```

## start using beautiful souap library

```python
first_header = soup.find("h2")
print(first_header)
print("------------------")
headers = soup.find_all("h2")
print(headers)
```

```python
first_header = soup.find(["h1", "h2"])
# first_header = soup.find_all(["h1", "h2"])
print(first_header)
```

```python
# you can pass in ettributes to the find and find_all
paragraph = soup.find_all("p" , attrs={"id":"paragraph-id"})
paragraph
```

```python
# you can nest find/find-all
body = soup.find("body")
div = body.find("div")
header = div.find("h1")
header
```

```python
# we can search for specific string in our find / find-all
import re
soup.find_all("p",string="Some bold text") # must insert complete text
soup.find_all("p",string=re.compile("Some")) # can insert some part of text
```

## select (css selector)
use css selectors

```python

content = soup.select("p")
content
```

```python
soup.body.prettify()
print(soup.select("div p")[0].prettify())
```

```python
paragraphs = soup.select("h2 ~ p")
for paragraph in paragraphs : 
    print(paragraph.select("i"))
```

## get different propertties of the HTML

```python
header = soup.find("h2")
print(header.string)
print("-----")
div = soup.find("div")
print(div.prettify())
print(div.get_text())
```
