### load libraries

```python
import requests 
from bs4 import BeautifulSoup as bs
```

## load our first page

```python
# load the website
r = requests.get("https://keithgalli.github.io/web-scraping/example.html")
# convert to beatiful object
soup = bs(r.content)
# print out our html
print(soup.prettify()) # or just print(soup)
```
